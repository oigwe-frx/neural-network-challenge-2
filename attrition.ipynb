## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

!pip install tensorflow
import tensorflow
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data
column_names = ["Education", "Age",
                "DistanceFromHome",
                "JobSatisfaction",
                "OverTime",
                "StockOptionLevel",
                "WorkLifeBalance",
                "YearsAtCompany",
                "YearsSinceLastPromotion",
                "NumCompaniesWorked"]

# Create X_df using your selected columns
X_df = pd.DataFrame(attrition_df, columns=column_names)
X_df.head()

# Show the data types for X_df
X_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=42)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
print(X_train.value_counts('OverTime'))

X_train['OverTime'] = X_train['OverTime'].replace({'Yes': 1, 'No': 0})
X_test['OverTime'] = X_test['OverTime'].replace({'Yes': 1, 'No': 0})

# Display the value counts of the 'OverTime' column in the training and testing sets
print("Value counts of 'OverTime' column in training set:")
print(X_train.value_counts('OverTime'))

print("\nValue counts of 'OverTime' column in testing set:")
print(X_test.value_counts('OverTime'))
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Convert the scaled data back to DataFrame to maintain column names
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
# Create a OneHotEncoder for the Department column
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse=False)

# Fit the encoder to the 'Department' column in y_df
encoder.fit(y_train[['Department']])

# Create two new variables by applying the encoder
# Apply the encoder to the 'Department' column in the training and testing sets
y_train_encoded_dept = encoder.transform(y_train[['Department']])
y_test_encoded_dept = encoder.transform(y_test[['Department']])

# Print the resulting arrays
print("Encoded 'Department' column in training set (as array):")
print(y_train_encoded_dept)

print("\nEncoded 'Department' column in testing set (as array):")
print(y_test_encoded_dept)
```

```python
# Create a OneHotEncoder for the Attrition column


# Fit the encoder to the training data
encoder.fit(y_train[['Attrition']])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_encoded_attr = encoder.transform(y_train[['Attrition']])
y_test_encoded_attr = encoder.transform(y_test[['Attrition']])

# Print the resulting arrays
print("Encoded 'Attrition' column in training set (as array):")
print(y_train_encoded_attr)

print("\nEncoded 'Attrition' column in testing set (as array):")
print(y_test_encoded_attr)

```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
X_train_scaled.shape[1]

# Create the input layer
input_layer = layers.Input(shape=(X_train_scaled.shape[1],))

# Create at least two shared layers
shared_layer1 = layers.Dense(64, activation='relu')(input_layer)
shared_layer2 = layers.Dense(128, activation='relu')(shared_layer1)
```

```python
# Create a branch for Department with a hidden layer and an output layer

# Create the hidden layer
department_hidden = layers.Dense(32, activation='relu')(shared_layer2)

# Create the output layer
department_output = layers.Dense(3, activation='softmax')(department_hidden)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_hidden = layers.Dense(32, activation='relu')(shared_layer2)

# Create the output layer
attrition_output = layers.Dense(2, activation='softmax')(attrition_hidden)
```

```python
# Create the model

model = Model(inputs=input_layer, outputs=[department_output, attrition_output])

# Compile the model
model.compile(optimizer='adam',
              loss={'dense_2': 'categorical_crossentropy', 'dense_4': 'categorical_crossentropy'},
              metrics={'dense_2': 'accuracy', 'dense_4': 'accuracy'})

# Summarize the model
```
